```python
# Final reconciliation tables
fact_break_table = "fact_reconciliation_break"
daily_table = "reconciliation_daily"
```

## Monthly close query


```python
# for summary view we can lookup reconciliation_daily

df = spark.sql(f"""
    SELECT business_date, operator_code,
           SUM(partner_amount_total)  AS partner_revenue,
           SUM(internal_amount_total) AS platform_revenue,
           SUM(variance)              AS net_variance,
           SUM(break_count)           AS total_breaks
    FROM   {daily_table}
    WHERE  business_date BETWEEN DATE '2026-05-01' AND DATE '2026-05-31'
    GROUP BY business_date, operator_code
    ORDER BY business_date, operator_code
""")

display(df)
```

```python
# for drill down detailed view we have to use fact_reconciliation_break

df = spark.sql(f"""
    SELECT partner_txn_id, internal_id, break_category,
           partner_amount, internal_amount, variance, source_rule
    FROM   {fact_break_table}
    WHERE  business_date = DATE '2026-05-20'
    ORDER BY break_category
""")

display(df)
```
